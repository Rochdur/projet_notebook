Ce déséquilibre reflète une forte hétérogénéité dans les niveaux de salaires et souligne l'importance d'utiliser des statistiques robustes, telles que la médiane.

Cette analyse révèle que les entreprises établies dans certains pays, comme Israël et les États-Unis, proposent en moyenne les salaires les plus élevés. Cette disparité géographique illustre les écarts dans la maturité des marchés de la data et le degré de spécialisation des postes.

```python
roles_100k = (
    df.groupby("job_title")["salary_in_usd"]
      .mean()
      .loc[lambda x: x > 100000]
      .sort_values(ascending=False)
)

roles_100k
```

Les postes dont le salaire moyen excède 100 000 USD se concentrent principalement sur des fonctions à haute responsabilité et nécessitant une expertise technique poussée. Les rôles de lead, d’architecte et de direction dans le domaine de la data figurent parmi les mieux payés, mettant en évidence l'importance de la spécialisation et des compétences en gestion au sein des professions liées à la data.

```python
remote_100 = df[df["remote_ratio"] == 100]

pays_remote = (
    remote_100["company_location"]
    .value_counts(normalize=True)
    .sort_values(ascending=False)
)

pays_remote
```

L'analyse révèle que les postes entièrement en télétravail se situent majoritairement aux États-Unis. Cette concentration significative témoigne d'une adoption plus poussée du télétravail dans les entreprises technologiques américaines, alors que les autres pays demeurent bien moins présents dans ces pays.

```python
df_remote_100 = df[df["remote_ratio"] == 100]

salaire_remote = df_remote_100["salary_in_usd"].mean()

mask_tech = df["job_title"].str.contains(
    "Data|Machine Learning|ML|Engineer|Scientist|AI",
    case=False,
    na=False
)

df_tech = df[mask_tech]
df_non_tech = df[~mask_tech]
print("Salaire moyen (tech) :", df_tech["salary_in_usd"].mean())
print("Salaire moyen (non-tech) :", df_non_tech["salary_in_usd"].mean())
print("Salaire moyen des employés à distance :", salaire_remote)
```

L'analyse révèle que les employés exerçant entièrement à distance bénéficient, en moyenne, de salaires élevés. De plus, les postes techniques se distinguent par une rémunération nettement supérieure à celle des postes non techniques, mettant en lumière l'importance des compétences technologiques et de la spécialisation dans la détermination des niveaux de salaire.

```python
bins = [0, 50000, 100000, 150000, 200000, float("inf")]
labels = ["<50k", "50k-100k", "100k-150k", "150k+", "200k+"]

df["salary_range"] = pd.cut(
    df["salary_in_usd"],
    bins=bins,
    labels=labels,
    right=False
)

print("Données avec fourchettes de salaires :")
df[["salary_in_usd", "salary_range"]].head()
```

Grâce à l'utilisation de la fonction pd.cut, les salaires ont été classés en intervalles, ce qui permet de convertir une variable quantitative en variable catégorielle. Cette méthode rend l'étude et la représentation des niveaux de salaire plus aisées.

```python
print("Pays avec les salaires moyens les plus élevés : employee_residence")

df.groupby("employee_residence")["salary_in_usd"] \
  .mean() \
  .sort_values(ascending=False) \
  .head()
```

L'étude des rémunérations moyennes selon le pays de résidence révèle d'importantes disparités géographiques. Dans certains pays, les moyennes peuvent être très élevées, influencées par quelques observations ou des postes de direction.

```python
correlation = df["remote_ratio"].corr(df["salary_in_usd"])
print("Corrélation télétravail / salaire :", correlation)
```

Il existe une très faible corrélation(Pearson), proche de zéro, entre le taux de travail à distance et la rémunération. Cela signifie que le travail à distance, en tant que tel, n'a pas d'influence marquante sur le niveau salarial. Il semble que les rémunérations soient plus liées au poste occupé, à l'expérience et à la localisation géographique

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x="remote_ratio",
    y="salary_in_usd"
)

plt.title("Distribution des salaires selon le taux de télétravail")
plt.xlabel("Taux de télétravail (%)")
plt.ylabel("Salaire (USD)")
plt.show()

```

La répartition des rémunérations est comparable, indépendamment du pourcentage de travail à distance. Les médianes et la variabilité restent similaires entre le travail en présentiel, le travail à distance partiel et le travail à distance complet, ce qui confirme que le télétravail n'est pas un élément crucial dans la détermination du niveau de rémunération.

### Ce que j'ai trouvé interessant à faire c'est de voir la moyenne des salaires par niveau de télétravail.

```python
df.groupby("remote_ratio")["salary_in_usd"].mean()
```

L'analyse les salaires moyens en fonction du degré de télétravail indique que les employés travaillant sur site et à distance totale ont des salaires moyens similaires, alors que le travail à distance partiel est lié à des rémunérations moyennes inférieures. Cela indique que le travail à distance n'est pas un élément décisif pour déterminer le niveau de rémunération.

```python
df.groupby("company_size")["salary_in_usd"].mean().sort_values(ascending=False)
```

L'analyse indique que les entreprises de taille intermédiaire proposent, en général, les salaires les plus importants, suivies par les grandes firmes, alors que les petites entités affichent des niveaux de rémunération inférieurs. Ainsi, la dimension de l'entreprise se révèle être un élément influençant le salaire, bien que d'autres facteurs aient aussi une importance considérable.

### Visualisation  des salaire moyen par taille d’entreprise

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x="company_size",
    y="salary_in_usd",
    estimator="mean"
)

plt.title("Salaire moyen par taille d'entreprise")
plt.xlabel("Taille de l'entreprise")
plt.ylabel("Salaire moyen (USD)")
plt.show()
```

Les entreprises de taille intermédiaire offrent les salaires les plus importants, suivies par les grandes entreprises, tandis que les petites entreprises tendent à offrir des rémunérations plus modestes.
Ainsi, la dimension de l'entreprise joue un rôle dans la détermination du salaire, mais n'est pas le seul élément explicatif.
